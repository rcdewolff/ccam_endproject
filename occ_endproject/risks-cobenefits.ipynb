```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Data
methods = ["OAE with OL", "Electrochemical CDR", "Membrane", "Microalgae", "Macroalgae"]
severity = [3, 2, 1, 2, 2]
unpredictability = [4, 4, 1, 3, 2]
scale = ["local", "local", "local", "local", "global"]
#colors = ['blue' if s == 'local' else 'red' for s in scale]

fig, ax = plt.subplots()

# Creating scatter points and labeling them
for i in range(len(methods)):
    ax.scatter(severity[i], unpredictability[i])#, color=colors[i])
    ax.text(severity[i], unpredictability[i], methods[i], fontsize=11, ha='right')

# Setting the x and y axis limits
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)

# Adding labels and title
ax.set_xlabel('Externalities (1 (Positive or no negative) - 5 (Very negative))')
ax.set_ylabel('Uncertainty (1 (Very predictable) - 5 (Very unpredictable))')
ax.set_title('Oceanic Carbon Capture Methods: Risk/Co-Benefits comparison')

# Creating a custom legend for 'local' and 'global'
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Local',
                          markerfacecolor='blue', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='Global',
                          markerfacecolor='red', markersize=10)]
#ax.legend(handles=legend_elements, loc='upper right')

# Show the plot with grid
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating dummy data
np.random.seed(0)  # Seed for reproducibility

# Creating dummy data for 5 methods
methods = ["OAE with OL", "Electrochemical CDR 1", "Electrochemical CDR 2", "Electrochemical CDR 3", "Membrane", "Microalgae", "Macroalgae", "Direct Air Capture"]
cost = [(90.32+192.69) / 2, (373+717) / 2, (583 + 1076) / 2, (2355 + 1076) / 2, 10000, (280+30000) / 2, (16+300) / 2, (150 + 1000) / 2] # Cost in dollars per tonne CO2
scalability = [582.9, 50, 60, (100 + 1000) / 2, 1, 8.67e3, (100+5e3) / 2, 980]  # Scalability in Mt CO2

# Generating dummy uncertainty ranges (standard deviation for simplicity)
cost_uncertainty = [(192.69 - 90.32) / 2, (717 - 373) / 2, (1076 - 583) / 2, (2355 - 1076) / 2, 0, (30000 - 280) / 2, (300 - 16) / 2, (1000 - 150) / 2] # Uncertainty in cost
scalability_uncertainty = [582.9 * 0.75, 0, 0, (1000 - 100) / 2, 0, 0, (5e3 - 100) / 2, 0] # Uncertainty in scalability

# Plotting
fig, ax = plt.subplots()

# Scatter plot with error bars for uncertainty
for i in range(len(methods)):
    ax.errorbar(cost[i], scalability[i], 
                 xerr=cost_uncertainty[i], yerr=scalability_uncertainty[i], 
                 fmt='o', capsize=5, label=methods[i])

# Adding labels and title
ax.set_xlabel('Cost ($ per tonne CO2)')
ax.set_ylabel('Scalability (Mt CO2 / year)')
ax.set_title('Cost vs Scalability of Oceanic Carbon Capture Methods')
ax.legend(loc='lower left')
ax.set_xscale('log')
ax.set_yscale('log')

# Show the plot with grid
plt.grid(True)
plt.tight_layout()
plt.show()
```
